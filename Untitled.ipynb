### graph의 종류
- 방향 그래프
- 무방향 그래프
- 가중치가 반영된 그래프
- 멀티그래프

```python
import networkx as nx
```

```python
#방향 그래프

G_directed = nx.DiGraph()

G_directed.add_edge('B','A')
G_directed.add_edge('B','C')
G_directed.add_edge('C','E')
G_directed.add_edge('C','F')
G_directed.add_edge('E','D')
G_directed.add_edge('G','F')
```

```python
nx.draw_networkx(G_directed)
```

```python
#무방향 그래프
```

```python
G_undirected = nx.Graph()

G_undirected.add_edge('B','A')
G_undirected.add_edge('B','C')
G_undirected.add_edge('C','E')
G_undirected.add_edge('C','F')
G_undirected.add_edge('E','D')
G_undirected.add_edge('G','F')
```

```python
nx.spring_layout(G_undirected)
nx.draw_networkx(G_undirected)
```

```python
#가중치가 반영된 그래프

G_weighted = nx.Graph()

G_weighted.add_edge('B','A', weight = 6)
G_weighted.add_edge('B','C', weight = 12)
G_weighted.add_edge('C','E', weight = 6)
G_weighted.add_edge('C','F', weight = 18)
G_weighted.add_edge('E','D', weight = 3)
G_weighted.add_edge('G','F', weight = 15)
```

```python
nx.circular_layout(G_weighted)
nx.draw_networkx(G_weighted)
```
